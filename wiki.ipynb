```python
from bs4 import BeautifulSoup
import requests
import tqdm
import json
```

```python
base = 'https://en.wikipedia.org/wiki/Billboard_Year-End_Hot_100_singles_of_'
```

```python
articles = []
for year in range(1959, 2018):
    articles.append(base + str(year))
```

```python
articles
```

```python
r = requests.get(articles[0])
```

```python
soup = BeautifulSoup(r.text, 'lxml') # Parse HTML
wiki_table = soup.find('table', {'class': 'wikitable sortable'})
```

```python
songs = []
for row in wiki_table.findAll('tr')[1:]:
    title, artist = row.findAll('td')[1:]
    link = title.a['href']
    r = requests.get('https://en.wikipedia.org' + link)
    soup = BeautifulSoup(r.text, 'lxml')
    genre = soup.find(title='Music genre')
    if genre is not None:
        genre = genre.parent.parent.find('td').text
        songs.append({ 'artist': artist.text, 'title': title.text, 'genre': genre.lower() })
```

```python
songs = []

for article in tqdm.tqdm(articles):
    article_request = requests.get(article)
    article_soup = BeautifulSoup(article_request.text, 'lxml')
    table = article_soup.find('table', { 'class': 'wikitable sortable' }).find('tbody')
    
    for row in table.findAll('tr'):
        tds = row.findAll('td')
        
        if len(tds) == 2:
            title, artist = tds
        elif len(tds) == 3:
            title, artist = tds[1:3]
        
        if title.a is not None:
            link = title.a['href']
            print(link)
            song_request = requests.get('https://en.wikipedia.org' + link)
            song_soup = BeautifulSoup(song_request.text, 'lxml')

            genre_th = song_soup.find(title='Music genre')
            if genre_th is not None:
                parent_td = genre_th.parent.parent.find('td')
                genre = None
                if parent_td.a is None:
                    genre = parent_td.text
                else:
                    genre = parent_td.a.text
                songs.append({ 'artist': artist.text, 'title': title.text, 'genre': genre.lower() })
```
